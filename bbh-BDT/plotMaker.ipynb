```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LogNorm

sns.set(font_scale=1.25, style='white')
```

```python
def fileparser(path, dlist, frac=0.5, sample=0):
    df = pd.read_csv(path)
    df.drop(columns=dlist, inplace=True)    
    n = len(df)
    weight = int(round(np.mean(np.abs(df['weight'].to_numpy())) * 3 * 1e6 * n))
    df['weight'] = df['weight']/np.abs(df['weight'])
    df_test = df.sample(frac=frac)
    df.drop(df_test.index, inplace=True)
    if sample != 0: df = df.sample(n=sample).reset_index(drop=True)
    return df, df_test, weight
```
